# 3 HOW TO REMOVE DUPLICATE DATA

# Remove duplicate data

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
DF_obj = DataFrame({'Roll': [1, 2, 3, 4, 5, 6, 7, 8, 3, 3, 5, 5, 5],
                   'Name': ['Akkal', 'Janak', 'Laxman', 'Dinesh', 'Amin', 'Bikash', 'Sumil', 'Kiran', 'Laxman', 'Laxman', 'Amin', 'Amin', 'Amin'],
                   'Marks': [10, 20, 30, 40, 50, 60, 70, 80, 30, 30, 50, 50, 50]})
```

```python
DF_obj
```

```python
DF_obj.duplicated()
```

```python
DF_obj.drop_duplicates()
```

```python
DF_obj.drop_duplicates(['Marks'])
```

```python
DF_obj
```

# 4 CONCATINATING AND TRANSFORMING DATA

# Concatinating Data

```python
DF_obj = pd.DataFrame(np.arange(36).reshape(6,6))
```

```python
DF_obj
```

```python
DF_obj_2 = pd.DataFrame(np.arange(15).reshape(5,3))
```

```python
DF_obj_2
```

```python
pd.concat([DF_obj,DF_obj_2], axis = 1)
```

```python
pd.concat([DF_obj,DF_obj_2])
```
