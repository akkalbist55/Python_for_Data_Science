# 3 HOW TO REMOVE DUPLICATE DATA

# Remove duplicate data

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
DF_obj
```

```python
DF_obj.duplicated()
```

```python
DF_obj.drop_duplicates()
```

```python
DF_obj.drop_duplicates(['Marks'])
```
